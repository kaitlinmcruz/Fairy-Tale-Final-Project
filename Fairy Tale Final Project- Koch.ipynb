In relation to a previous project focused on analyzing the interactions between main characters and anthropomorphized characters within a small subset corpus of Grimm's fairy tales, this project seeks a deeper understanding of network connections and linguistic similarity between 62 of Grimm's fairy tales as found on the Project Gutenberg website. Previously, the corpus included only four texts selected from the Grimm's works, starting with well-known Cinderella, or Aschenputtel, continuing on with Little Red Riding Hood, or Rotkäppchen, and moving to the lesser well-known works of The Seven Ravens, or Die sieben Raben, and ending with Snow-White and Rose-Red, or Schneeweißchen und Rosenroth. The original idea was to show that it is possible to determine character relationality through text mining and distance reading measures with visual plot charts. The conclusion of this project stated, "the inserted graphs of the word trends in the stories, taken from Voyant after uploading the corpus that included one tale at a time, show how the characters dance in and out of the story and around one another. What is not quite clear here is how the characters specifically interact with one another, especially in terms of sentimentally charged words within the text. With these aspects in mind, the scope of this project has been expanded to include a significantly larger corpus of texts, this time in English as found on Gutenberg. Since the corpus is now in English, lemmatizing, part of speech tagging, and sentiment analysis could be and may be applied as well as document similarity measures that convert simple frequencies of terms within texts compared to the overall frequencies in order to create a perceptable network of tales. 

To start, this project seeks to create a network similarity graph of all texts within the corpus. In order to do this, the first steps will be to load the corpus, compute the frequencies of terms as raw counts within each text as well as the inverse frequencies within the corpus of documents to form a ratio of a given term's 'importance' or rather uniqueness therein, create a document similarity matrix from this output to determine relationality between the texts, plot a scatter plot of the TF-IDF measures that will show linguistically similar clusters of texts, and a dendrogram which will create a hierarchy of the clusters of text, proving the clustering provided in the previous scatter plot.

The IDF portion is how important each word is within the files, i.e. how often it occurs compared to other terms. The first step to getting a sparse matrix of the term frequencies is to create an array of all of the texts we want to be read and tabulated. 

The CountVectorizer library is instantiated counts the tokens within the corpus and converts those values into the matrix. The output is called a document term matrix (DTM), and in our case it will be sparse because all texts are linguistically different, and therefore do not all contain the same terms, and usually not the same number of times. # Sparse Matrix of Term Frequencies

Assuming that no error means this worked properly, we can continue on to the sparse matrix functionality, where the terms in the documents are tokenized, splitting them apart into a bag of words, and then their frequencies are counted. On the left side of the table we will see a list of numbers 0-61, these numbers correspond to each text file from Ashputtel[0] to Tom Thumb[61]
We will want to remove stopwords and punctuation because of their lack of importance to the overall plot of the tales.

## Document Term Matrix: "TF"

```python
allWords = set([word.lower() for word in nltk.word_tokenize(" ".join(fairyTales)) if any([c for c in word if c.isalpha()])])
allRawFreqs = [nltk.FreqDist(nltk.word_tokenize(fairyTales.lower())) for fairyTales in fairyTales]
pd.DataFrame(allRawFreqs).fillna(0)
```

The 62 rows correspond again to each fairy tale, and the 5223 columns represent the number of tokens found within the text. Here, the term "sparse matrix" comes into play insomuch as there are a significant number of cells which contain no information, since each text is its own entity and therefore will not necessarily include the same words as other texts. The 0s were input in the code above where .fillna() can be seen. This means that when the function encounters no instance of a word within a document, normally it would return n/a or NaN, but we have told it to input a 0 so that the next function can be excecuted. This is the TF part of the TF-IDF. Meaning that from here we can go on to do the second step, getting the IDF. 

## Inverse Document Frequency Matrix: "IDF"

The IDF, inverse document frequency, is the log of the total number of documents divided by the number of documents that contain a given term. There is a specific vectorizer library that will calculate these values automatically instead of having to do them manually, which will save time. 

```python
from sklearn.feature_extraction.text import TfidfVectorizer as TFV
fairydtm = TFV().fit_transform(fairyTales)#this tokenizes the strings
pd.DataFrame(fairydtm.toarray())
```

This dataframe is output as a sparse matrix which allows for a table of data where cells may not have any values. The 4759 columns refer to the number of unique tokens/terms/words within the corpus. Therefore, the columns are the terms, which have had their actual terms stripped away through the simplified output calculation process completed above. However, knowing the terms is not the most important part of the analysis. What is important is the next step which is comparing these values and documents to one another, which will give an overall "distance" between them based on the sparse matrix information. 

## Document Similarity

In order to continue on to the next step, the TF-IDF values need to be converted into geometric space which will allow for visual graphing and clustering of the documents and how they relate to one another, i.e. a related network. Each document then will be expressed as an aggregate of the points on a cartesian graph relating to the values output in the matrix. Then, each document will be expressed as a vector (line) with a length and a magnitude from the origin (0,0). We cannot compare the lines for each document here, but we can compare their angles, which is known as cosine similarity. Once again, there is a library that contains the cosine similarity measures which will create a new matrix of the distances between documents similar to distances between physical geospatial locations.

### Cosine Similarity

The following function will import the data from the distances matrix above and conver them into an array of lines, which can then have the angles between them measured and charted as a two dimensional similarity graph.

```python
from sklearn.metrics.pairwise import cosine_similarity as cos
distances = 1 - cos(fairydtm)
distances
```

Since these distances are in terms of multiple dimensions, they need to be converted into a two dimensional arena for graphing purposes. Again this can be completed easily using the multidimensional scaling measure which takes the n-dimensionality of the matrix and converts it into a two dimensional matrix with x and y coordinates for plotting.

```python
from sklearn.manifold import MDS
mds = MDS(dissimilarity="precomputed", random_state=1)
positions = mds.fit_transform(distances)
positions
```

There are now 62 elements which correspond to the 62 texts within the Fairy Tale corpus from Project Gutenberg, and each text has a corresponding x and y value. From here the next step is to plot these values and see which tales are computationally (linguistically) most similar, and whether there are clusters or families of texts. We create a name object for the scatter plot to associate plotted points to the file names of each text. The figure size can be scaled, which allows for easier readability, which otherwise could be impossible due to the nature of the clusters. 

```python
names = corpus_root.fileids()
import matplotlib.pyplot as plt
xvalues = positions[:,0] #this extracts only the left column of the array for the x axis
yvalues = positions[:,1] #this does the same thing for the y axis
plt.figure(figsize=(10,10)) #this makes the graph larger
for x, y, names in zip(xvalues, yvalues, names):
    plt.scatter(x, y)
    plt.text(x, y, names.replace(".txt", " "))
plt.show()
```
